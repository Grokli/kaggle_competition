```python
import numpy as np
import pandas as pd

train_df = pd.read_csv('train_V2.csv')
test_df = pd.read_csv('test_V2.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_n = len(train_df)
all_df = pd.concat([train_df,test_df], ignore_index=True ,sort=False)
```

```python
all_df.tail()
```

```python
all_df.info()
```

```python
all_df['group_people_num']=all_df.groupby('groupId').Id.transform('count')
```

```python
all_df['machId_num'] = all_df.groupby('matchId').Id.transform('count')
```

```python
all_df.damageDealt = all_df.damageDealt.clip(0, all_df.damageDealt.quantile(q=0.8))
```

```python
all_df.killPlace.replace(101, 100, inplace=True)
```

```python
all_df.loc[(all_df.rankPoints!=-1) & (all_df.killPoints==0), 'killPoints'] = np.nan
```

```python
all_df.killPoints.fillna(all_df.killPoints.mode()[0], inplace=True)
```

```python
all_df.longestKill = all_df.longestKill.clip(0, all_df.longestKill.quantile(q=0.8))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stds = StandardScaler()
```

```python
train_df_matchDuration = stds.fit_transform(train_df.matchDuration.values.reshape(-1,1))
```

```python
test_df_matchDuration = stds.transform(test_df.matchDuration.values.reshape(-1,1))
```

```python
all_df.matchDuration = np.vstack([train_df_matchDuration, test_df_matchDuration])
```

```python
dummies = ['matchType']
```

```python
all_df['different'] = 0
```

```python
all_df.loc[all_df.maxPlace!=all_df.numGroups, 'different'] = 1
```

```python
train_df_rideDistance = stds.fit_transform(train_df.rideDistance.values.reshape(-1,1))
test_df_rideDistance = stds.transform(test_df.rideDistance.values.reshape(-1,1))
all_df.rideDistance = np.vstack([train_df_rideDistance, test_df_rideDistance])
```

```python
train_df_swimDistance = stds.fit_transform(train_df.swimDistance.values.reshape(-1,1))
test_df_swimDistance = stds.transform(test_df.swimDistance.values.reshape(-1,1))
all_df.swimDistance = np.vstack([train_df_swimDistance, test_df_swimDistance])
```

```python
train_df_walkDistance = stds.fit_transform(train_df.walkDistance.values.reshape(-1,1))
test_df_walkDistance = stds.transform(test_df.walkDistance.values.reshape(-1,1))
all_df.walkDistance = np.vstack([train_df_walkDistance, test_df_walkDistance])
```

```python
all_df.loc[(all_df.rankPoints!=-1) & (all_df.winPoints==0), 'winPoints'] = np.nan
all_df.winPoints.fillna(all_df.winPoints.mode()[0], inplace=True)
```

```python
all_df.loc[all_df.rankPoints==-1, 'rankPoints'] = np.nan
all_df.rankPoints.fillna(all_df.rankPoints.mode()[0], inplace=True)
```

```python
all_df = pd.get_dummies(all_df, columns=dummies)
```

```python
all_df.drop(['Id','groupId', 'matchId'], axis=1, inplace=True)
```

```python
train = all_df[:train_n]
test = all_df[train_n:]
train.dropna(inplace=True)
train_y = train['winPlacePerc']
train = train.drop('winPlacePerc', axis=1)
test = test.drop('winPlacePerc', axis=1)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, train_y, random_state=42)
rd = Ridge()
rd.fit(X_train, y_train)
```

```python
rd.score(X_test, y_test)
```
